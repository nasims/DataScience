### Read and save dataset as a string:

```python
df = json.load(open('./data/world_bank_projects.json')) 
```

### Create table from nested elements in mjtheme_namecode column:

```python
df = json_normalize(df, 'mjtheme_namecode')
df.head(10)
```

### Find the top 10 major project themes by code:

```python
df.code.value_counts().head(10)
```

### Drop duplicate columns to create a map for code and name:

```python
df_map = df.drop_duplicates().loc[df.name!='']
df_map.head(20)
```

```python
df_map = df_map.set_index('code').to_dict().get('name')
df_map
```

### In the df, some entries have only the code and the name is missing. We want to create a dataframe with the missing names filled in. First df is sorted by 'code' and 'name', and then the missing 'name' values is filled with NaN. 

```python
df = df.sort_values(['code', 'name'])
df.loc[df.name=='','name']=np.nan
df.loc[df.name=='','code']=np.nan
```

### In the df, replace NaN values with the values from the df_map dataframe:

```python
df3 = df.copy()
df3.name = df3.name.fillna(df3.code.apply(lambda x: df_map.get(x)))
df3.name.value_counts().head(10)
```

## Second Method:
### Since the values are sorted, the NaN values can be filled using backward filling method, bfill:

```python
df2 = df.fillna(method='bfill')
```

### Now that all the missing values in 'name' column are filled, we can find the top ten major project themes by name:

```python
df2.name.value_counts().head(10)
```
